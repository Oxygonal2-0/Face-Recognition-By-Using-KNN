```python
import cv2
```

```python
# Read an Image

img = cv2.imread('Codechef.jpg') #RGB image converted to BGR
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
plt.show()
plt.imshow(newImg)
plt.show()
```

```python
# Simple Program to read and show an image by using opencv
cv2.imshow("Codechef Logo", img)    # It automatically convert BGR to RGB
cv2.waitKey(0)   # Program hold for infinite time
cv2.destroyAllWindows()
```
